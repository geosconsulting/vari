```python
import pandas as pd
from great_tables import GT, md, html
import matplotlib.pyplot as plt
from plottable import Table
```

```python
df = pd.read_csv("../data/negozio/1-gen-giu-24.csv", encoding="windows-1252", sep=",")
```

```python
my_table = (
    GT(df.head(4))
    .tab_header(
        title="Vendite",
        subtitle="Lista delle vendite",
    )
    .tab_source_note(md("**Data source**: vendite negozio gestionale"))    
)
```

```python
fig, ax = plt.subplots(figsize=(6, 3))

tab = Table(df.sample(10), row_dividers=False, odd_row_color="#1bfab0", even_row_color="#ddf318")

plt.show()
```

```python
def custom_table(styler):
    styler.background_gradient(cmap="Reds", axis=None)

    # center text and increase font size
    styler.applymap(lambda x: "text-align: center; font-size: 14px;")
    return styler


df.style.pipe(custom_table)
```
